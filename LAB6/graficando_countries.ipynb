```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv("countries.csv")
data 
```

```python
brazil = data[data.country =="Brazil"]
brazil
```

```python
argentina = data[data.country =="Argentina"]
argentina
```

```python
plt.plot(brazil.year, brazil.population/10**6)
plt.plot(argentina.year, argentina.population/10**6)
plt.legend(["Brazil", "Argentina"])
plt.xlabel("Año")
plt.ylabel("Poblacion (millones)")
plt.show()
```
